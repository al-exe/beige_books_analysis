```python
employ = beige[beige['Sector Heading'].isin(['Unemployment','Employment','Employment and Unemployment'])]
```

```python
#EMPLOYEMENT across all districts, all years
#making words readable to wordcloud

long = []
for lst in employ['keywords']:
    lst = eval(lst)
    long = long + lst

string = ''
for i in long:
    string = string + ' ' + i


#visualizng commonly spoken words under employment section
wordcloud = WordCloud().generate(string)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(string)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from textblob import TextBlob
```

```python
import seaborn as sns
```

```python
beige2 = beige2[beige2.Date != '1-1-,']
beige2 = beige2[beige2.Date != '11-1-,']
beige2 = beige2[beige2.Date != '4-24-200']
beige2 = beige2[beige2.Date != '9--,']
beige2 = beige2[beige2.Date != '10-2-,']



for i in beige2.Date:
    print(i)
    pd.to_datetime(i)
```

```python
years = [i.year for i in pd.to_datetime(beige2.Date)]
beige2['year']=years
```

```python
tbl00 = beige2[beige2.year.isin(np.arange(2000,2010))]
```

```python
tbl80s = beige2[beige2.year.isin(np.arange(1980,1991))]
```

```python
len(tbl80s)
```

```python
sent2 = []
for p in tbl00.clean:
    text = TextBlob(p)
    sent2 = np.append(sent2,text.sentiment.polarity)
sent2
```

```python
tbl00['sentiments'] = sent2
```

```python
tbl00.head()
```

```python
sentiments = []
for p in tbl80s.clean:
    text = TextBlob(p)
    sentiments = np.append(sentiments,text.sentiment.polarity)
sentiments
```

```python
sentiments = list(sentiments)
```

```python
tbl80s['sentiments'] = sentiments
```

```python
tbl80s.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(14, 5)
sns.boxplot(x=tbl80s.District, y=tbl80s.sentiments)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(14, 5)
sns.boxplot(x=tbl00.District, y=tbl00.sentiments)
plt.show()
```

The sentiment of records during the years of 1980-1990 and 2000-2010 are both very close to 0, only very slightly positive as expected.
