## Solution to extracting 10 keywords:

```python
ids = []
for lst in bowten:
    if len(lst)==10:
        lst_id = [lst[i][0] for i in range(10)]
    else:
        lst_id = [lst[i][0] for i in range(len(lst))]
    ids.append(lst_id)

ids
```

```python
ids
```

```python
keywords = [[dictionary[i] for i in ids[j]] for j in range(len(ids))]
```

```python
keywords
```

```python
# Adding keywords to dataframe
beige['keywords'] = keywords
```

# Applying to new dataframe with ONLY nouns
---

*run when new noun_keywords.csv done*

```python
import pandas as pd
import gensim
from gensim import corpora
```

```python
beige = pd.read_csv('noun_keywords.csv')
corpus = [eval(beige['cleaner'][i]) for i in range(len(beige))]
type(corpus[:3][0])
```

```python
dictionary = corpora.Dictionary(corpus)

#bow is a list of lists, each list containing the bag of words of one record
bow = [dictionary.doc2bow(doc) for doc in corpus]

#sorts every single list of bag of words
#bowsort doesn't return anything
bowsort = [bow[i].sort(key= lambda tup:tup[1], reverse=True) for i in range(len(bow))]

#creating list of top 10 keywords per record
bowten=[i[:10] for i in bow]
```

```python
ids = []
for lst in bowten:
    if len(lst)==10:
        lst_id = [lst[i][0] for i in range(10)]
    else:
        lst_id = [lst[i][0] for i in range(len(lst))]
    ids.append(lst_id)

ids
```

```python
keywords = [[dictionary[i] for i in ids[j]] for j in range(len(ids))]

keywords[:5]
```

```python
# Adding keywords to dataframe
beige['keywords'] = keywords
```

```python
#download en on spaCy
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
def tablefier(parsed_text):
    toks_tab = pd.DataFrame()
    toks_tab['Word'] = [word.text for word in parsed_text]
    toks_tab['POS'] = [word.pos_ for word in parsed_text]
    return toks_tab
```

```python
parsed = nlp(beige['nostop_stem'][0])
```

```python
gram = tablefier(parsed)
gram= gram[gram['POS'] != 'PUNCT']
gram=gram[gram['POS']!='ADP']
```
